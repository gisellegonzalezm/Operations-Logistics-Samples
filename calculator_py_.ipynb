```python
def calculate_cif(row):

    if row["Incoterm"] == "CIF":return row["Product_Value_USD"]

    elif row["Incoterm"] == "FOB":
        return (row["Product_Value_USD"]
            + row["Freight_USD"]
            + row["Insurance_USD"])

    elif row["Incoterm"] == "EXW":
        return (row["Product_Value_USD"]
            + row["Inland_Transport_USD"]
            + row["Freight_USD"]
            + row["Insurance_USD"])
```

```python
df["CIF_Value_USD"] = df.apply(calculate_cif, axis=1)
```

```python
df[ ["Shipment_ID", "Incoterm", "Product_Value_USD","CIF_Value_USD"]]
```

Calcular el arancel

```python
df["Duty_USD"] = df["CIF_Value_USD"] * (df["Duty_Rate"] / 100)
```

```python
df[["Shipment_ID","CIF_Value_USD","Duty_Rate","Duty_USD"]]
```

Calcular el IVA

```python
df["VAT_USD"] = (df["CIF_Value_USD"] + df["Duty_USD"]) * (df["VAT_Rate"] / 100)
```

```python
df[["Shipment_ID","VAT_USD"]]
```

Calcular el costo total

```python
df["Total_Import_Cost_USD"] = (df["CIF_Value_USD"]+ df["Duty_USD"]+ df["VAT_USD"])
```

```python
df[[ "Shipment_ID","Incoterm","CIF_Value_USD","Duty_USD","VAT_USD","Total_Import_Cost_USD"]]
```
